```python
#Geometry Graph - Clustering (500x10)

import csv
import sklearn.svm
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

dataset = []
index = []

with open('NodeLasVegas.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    df = pd.DataFrame(reader)
    for i in range(len(df)):
        if df[1][i] != 'lat' and df[2][i] != 'lon':
            dataset.append([float(df[1][i]), float(df[2][i])])
            # index.append(i)
    print(len(df))
            
with open('LasVegasNode1.csv', 'r', encoding='utf-8') as f2:
    reader2 = csv.reader(f2)
    df2 = pd.DataFrame(reader2)
    for i in range(len(df2)):
        if df2[0][i] != 'lat' and df2[1][i] != 'lon':
            dataset.append([float(df2[0][i]), float(df2[1][i])])
    print(len(df2))
        
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=round(len(dataset)/500))
kmeans.fit(dataset)
labels = kmeans.labels_

counter = Counter(labels)
print(counter)
print(len(dataset))
df = pd.DataFrame(labels)
df.to_csv('LasVegasClustering.csv', index=False, header=False, encoding='utf-8')
```
